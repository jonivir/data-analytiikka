```python
df['tyytyväisyys']=df[['johto', 'työtov', 'työymp', 'palkkat', 'työteht']].mean(axis=1)
df.head(6)
```

```python
df['käyttö']=df[['työterv', 'lomaosa', 'kuntosa', 'hieroja']].sum(axis=1)
df.head(6)
```
