```python
import pandas as pd
import seaborn as sns
```

```python
#Luetaan kunnat.xlsx muuttujaan df ja tulostetaan 5 ensimmäistä riviä.

df=pd.read_excel('kunnat.xlsx', sheet_name='Kuntien avainluvut')
df.head()
```

```python
df.count()
```

```python
#Uudelleennimetään tarvittavat muuttujat.

df.rename(columns={'Kuntien välinen muuttovoitto/-tappio, henkilöä, 2017':'muuttovt', 'Väkiluku, 2017':'väkiluku'}, inplace=True)
df.columns
```

```python
#Valitaan kunnat joissa muuttotappiota ja järjestetään väkilukun mukaan laskevaan järjestykseen ja näytetään 10 ensimmäistä.

df[(df['muuttovt']<0)].sort_values(by='väkiluku', ascending=False).head(10)
```

```python
#Suurimmat muuttovoittoa saaneet kunnat
df2 = df[(df['muuttovt']>0)].sort_values(by='väkiluku', ascending=False).head(10)
df2.sort_values(by='väkiluku', ascending=False).head(10)
```

```python
df.plot.scatter('väkiluku', 'muuttovt')
```

```python
df.corr().style.format('{:.2f}')
```

```python
df.rename(columns={'Taajama-aste, %, 2017':'taajamaaste', 'Työllisyysaste, %, 2016' : 'työllisyysaste', 
                   'Korkea-asteen tutkinnon suorittaneiden osuus 15 vuotta täyttäneistä, %, 2017' : 'korkeakoulutetut',
                  'Alkutuotannon työpaikkojen osuus, %, 2016' : 'alkutuotanto', 
                   'Vähintään toisen asteen tutkinnon suorittaneiden osuus 15 vuotta täyttäneistä, %, 2017' : 'toinenaste',
                  'Yli 64-vuotiaiden osuus väestöstä, %, 2017' : 'yli64vuotiaat'}, inplace=True)
correlation_matrix = df[['alkutuotanto', 'työllisyysaste', 'korkeakoulutetut', 'muuttovt', 'väkiluku', 'taajamaaste', 'toinenaste', 
                        'yli64vuotiaat']].corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
## Taajama-asteen ja alkutuotannon työpaikkojen osuuden negatiivinen korrelaatio hajontakuviona
df.plot.scatter('taajamaaste', 'alkutuotanto')
df.plot.scatter('toinenaste', 'yli64vuotiaat')
```
