```python
import pandas as pd
```

```python
#Luetaan kunnat.xlsx muuttujaan df ja tulostetaan 5 ensimmäistä riviä.

df=pd.read_excel('kunnat.xlsx', sheet_name='Kuntien avainluvut')
df.head()
```

```python
df.count()
```

```python
#Uudelleennimetään tarvittavat muuttujat.

df.rename(columns={'Kuntien välinen muuttovoitto/-tappio, henkilöä, 2017':'muuttovt', 'Väkiluku, 2017':'väkiluku'}, inplace=True)
df.columns
```

```python
#Valitaan kunnat joissa muuttotappiota ja järjestetään väkilukun mukaan laskevaan järjestykseen ja näytetään 10 ensimmäistä.

df[(df['muuttovt']<0)].sort_values(by='väkiluku', ascending=False).head(10)
```
